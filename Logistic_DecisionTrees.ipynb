# Question 6 - Encode all categorical variables to numerical (0.5 points)

Take the list of categorical attributes(from the above result) and convert them into neumerical variables. After that, print the head of dataframe and check the values.

**Hint:** check **sklearn LabelEncoder()**

#### Answer:

```python
std_obj = std.select_dtypes(include=np.object)
#std_obj
from sklearn import preprocessing
# Create a label (category) encoder object
# Fit the encoder to the pandas column
#std_obj.apply(le.fit_transform)
for c in std.columns:
    if std[c].dtype == 'object':
        le = preprocessing.LabelEncoder()
        le.fit(list(std[c].values))
        std[c] = le.transform(list(std[c].values))
std.head()
#std.dtypes
```

### Divide the dataset into feature set and target set

```python
X=bc.drop(['Class'],axis=1)
y=bc["Class"]
```

### Divide the Training and Test sets in 70:30 

```python
# splitting data into training and test set for independent attributes
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=1)
X_train.shape
```

### Model 1 : Implement a default Decision Tree model with no attributes

```python
from sklearn.tree import DecisionTreeClassifier
reg_dt_model = DecisionTreeClassifier(criterion='entropy', max_depth=7)
reg_dt_model.fit(X_train,y_train)
```

### Check the feature importances of the Decision Tree model

```python
print (pd.DataFrame(reg_dt_model.feature_importances_, columns = ["Imp"], index = X_train.columns))
```

### Check the training & test accuracy of model 1 

```python
# calculate accuracy measures and confusion matrix
from sklearn import metrics
y_predict = reg_dt_model.predict(X_test)
print(reg_dt_model.score(X_train , y_train))
print(reg_dt_model.score(X_test , y_test))
```

```python
#print(metrics.confusion_matrix(y_test, y_predict))
```

### Model 2: Implement a decision tree with depth of the tree = 10 and leaf nodes = 5

```python
reg_dt_model2 = DecisionTreeClassifier(criterion='entropy', max_depth=10, max_leaf_nodes=5)
reg_dt_model2.fit(X_train,y_train)
```

### Check the training & test accuracy of model 2

```python
y_predict2 = reg_dt_model2.predict(X_test)
print(reg_dt_model2.score(X_train , y_train))
print(reg_dt_model2.score(X_test , y_test))
```

### Print the confusion matrix of model 1 and model 2
