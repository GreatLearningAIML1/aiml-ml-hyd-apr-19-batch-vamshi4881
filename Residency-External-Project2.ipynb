```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro,wilcoxon
from statsmodels.stats.power import ttest_power
```

```python
sales_df = pd.read_csv('sales_people.csv',delimiter=',')
#sales_df.head()
df = pd.DataFrame(pd.read_csv('sales_people.csv'))
df
```

```python
# What conclusion does the test (p-value) lead to? (2.5 points)
# Suppose it has been calculated that in order for Titan to break even, the average output must increase by Â£5000 in the scheme compared to the old scheme. 
# If this figure is alternative hypothesis, what is:
# The probability of a type 1 error? (2.5 points)
# What is the p- value of the hypothesis test if we test for a difference of $5000? (10 points)
# Power of the test (5 points)
# two-sample t-test
# null hypothesis: the new scheme has significant outputs on sales
# this test assumes the two groups have the same variance...
```

```python
# Find the mean of old scheme and new scheme column. (5 points)

# mean of the old_scheme and new_scheme column
old_scheme_mean = df.loc[:,"old_scheme"].mean()
new_scheme_mean = df.loc[:,"new_scheme"].mean()

old_scheme = np.array(sales_df[['old_scheme']])
new_scheme =np.array( sales_df[['new_scheme']])
print(old_scheme_mean)
print(new_scheme_mean)
```

```python
# Use the five percent significance test over the data to determine the p value to check new scheme 
# has significantly raised outputs? (10 points)

t_statistic, p_value = ttest_ind(old_scheme, new_scheme)
print(t_statistic, p_value)
```

```python
t_statistic, p_value = ttest_1samp(new_scheme - old_scheme, 0)
print(t_statistic, p_value)
```

```python
# p_value > 0.05 => null hypothesis is accepted:
# they have the significant output on the sales at the 5% significance level
```

```python
# What conclusion does the test (p-value) lead to?
# p < 0.05 => alternative hypothesis:
# the difference in mean is not equal to 0
print ("p-value=", p_value)
if(p_value>0.05):
    print("null hypothesis is accepted")
```
